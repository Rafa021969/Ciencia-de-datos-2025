 
 ESTUDIO DE BOLSA AMERICANA EN LOS ULTIMOS 10 AÑOS
 
 
 Las bolsas de valores cayeron en 2020 principalmente debido a la pandemia de COVID-19. A medida que el virus se propagaba rapidamente por todo el mundo, los gobiernos implementaron medidas de confinamiento, cierre de negocios y restricciones de viajes, lo que afecto gravemente a la economía global. Esto genero un aumento en la incertidumbre y el miedo, lo que llevo a los inversionistas a vender sus activos en masa.
 
 
 
 La caída de las bolsas americanas en 2022 y 2023 fue el resultado de una confluencia de factores como el aumento de la inflacion, las subidas de tasas de interes, los efectos del conflicto en Ucrania, la desaceleración economica global y la crisis energetica, que generaron un entorno de incertidumbre y volatilidad en los mercados.

 la caída de las bolsas en 2025 es el resultado de una combinacion de políticas proteccionistas(aranceles en el nuevo gobierno del presidente de Estados Unidos), tensiones geopolíticas, incertidumbre económica y pérdida de confianza en los activos financieros tradicionales. 

```python


# Importamos las librerías necesarias.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ruta de los archivos CSV
ruta_sp500_csv = 'sp500historico.csv'
ruta_nasdaq_csv = 'nasdaqhistorico.csv'

# Leemos los archivos CSV
df_sp500 = pd.read_csv(ruta_sp500_csv, sep=',')
df_nasdaq = pd.read_csv(ruta_nasdaq_csv, sep=',')

# Verificamos las columnas y los primeros registros
print(df_sp500.head())
print(df_nasdaq.head())

# Verificamos la informacion del DataFrame
print(df_sp500.info())  
print(df_nasdaq.info())  

# Aseguramos que la columna 'Fecha' este en formato de fecha
df_sp500['Fecha'] = pd.to_datetime(df_sp500['Fecha'])
df_nasdaq['Fecha'] = pd.to_datetime(df_nasdaq['Fecha'])

# Establecemos la columna de fecha como indice
df_sp500.set_index('Fecha', inplace=True)
df_nasdaq.set_index('Fecha', inplace=True)

# Verificamos que no haya valores nulos en las columnas relevantes antes de graficar
df_sp500 = df_sp500.dropna(subset=['Cerrar/último'])
df_nasdaq = df_nasdaq.dropna(subset=['Cerrar/último'])

# Graficamos el precio de cierre del indice S&P 500
plt.figure(figsize=(10, 6))
plt.plot(df_sp500['Cerrar/último'], label='Precio de Cierre S&P 500')
plt.title('Precio de Cierre del S&P 500')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.legend()
plt.grid(True)
plt.show()

# Graficamos el precio de cierre del Nasdaq
plt.figure(figsize=(10, 6))
plt.plot(df_nasdaq['Cerrar/último'], label='Precio de Cierre Nasdaq', color='orange')
plt.title('Precio de Cierre del Nasdaq')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.legend()
plt.grid(True)
plt.show()


# Graficamos ambos precios de cierre en un solo grafico 
plt.figure(figsize=(12, 6))
plt.plot(df_sp500['Cerrar/último'], label='S&P 500')
plt.plot(df_nasdaq['Cerrar/último'], label='Nasdaq', color='orange')

plt.title('Precio de Cierre: S&P 500, Nasdaq y Netflix')

plt.title('Precio de Cierre: S&P 500 vs Nasdaq')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.legend()
plt.grid(True)
plt.show()


# Segunda version.
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

axs[0].plot(df_sp500['Cerrar/último'], label='S&P 500')
axs[0].set_title('Precio de Cierre del S&P 500')
axs[0].set_ylabel('Precio de Cierre')
axs[0].legend()
axs[0].grid(True)

axs[1].plot(df_nasdaq['Cerrar/último'], label='Nasdaq', color='orange')
axs[1].set_title('Precio de Cierre del Nasdaq')
axs[1].set_xlabel('Fecha')
axs[1].set_ylabel('Precio de Cierre')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()




# Calculamos la media, mediana y moda del precio de cierre del S&P 500
media = df_sp500['Cerrar/último'].mean()
mediana = df_sp500['Cerrar/último'].median()
moda = df_sp500['Cerrar/último'].mode()[0]  # La moda podria tener mas de un valor, tomamos el primero



# Imprimimos los resultados
print(f"Media del precio de cierre: {media}")
print(f"Mediana del precio de cierre: {mediana}")
print(f"Moda del precio de cierre: {moda}")


# Agregamos las lineas verticales para la media, mediana y moda
plt.axhline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axhline(mediana, color='green', linestyle='--', label=f'Mediana: {mediana:.2f}')
plt.axhline(moda, color='blue', linestyle='--', label=f'Moda: {moda:.2f}')

# Añadimos el titulo, las etiquetas y la leyenda del indice S&P 500
plt.title('Precio de Cierre del S&P 500')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.legend()
plt.grid(True)

# Mostramos el grafico
plt.show()

# Creamos un nuevo grafico para mostrar la media, mediana y moda
# Creamos la figura y los ejes para graficar
plt.figure(figsize=(10, 6))

# Graficamos el precio de cierre
plt.plot(df_sp500['Cerrar/último'], label='Precio de Cierre S&P 500')

# Agregamos las líneas horizontales para la media, mediana y moda
plt.axhline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axhline(mediana, color='green', linestyle='--', label=f'Mediana: {mediana:.2f}')
plt.axhline(moda, color='blue', linestyle='--', label=f'Moda: {moda:.2f}')

# Añadimos el titulo, las etiquetas y la leyenda
plt.title('Precio de Cierre del S&P 500 con Media, Mediana y Moda')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.legend()
plt.grid(True)

# Mostramos el grafico
plt.show()



# Imprimimos los resultados de las estadísticas
print(f"Media del precio de cierre: {media}")
print(f"Mediana del precio de cierre: {mediana}")
print(f"Moda del precio de cierre: {moda}")




# Creamos un nuevo DataFrame para almacenar las estadísticas COVID-19 - De barras apiladas - mes a mes
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar los datos
df = pd.read_csv('covid19.csv')

# Asegurarse de que la columna 'date' esté en formato datetime
df['date'] = pd.to_datetime(df['date'])

# Crear columna de año-mes
df['year_month'] = df['date'].dt.to_period('M')

# Agrupar por año-mes y sumar los casos y muertes
monthly_data = df.groupby('year_month')[['cases', 'deaths']].sum().reset_index()
monthly_data['year_month'] = monthly_data['year_month'].astype(str)

# Crear gráfico de barras
x = np.arange(len(monthly_data['year_month']))
width = 0.4

plt.figure(figsize=(14, 7))
plt.bar(x - width/2, monthly_data['cases'], width, label='Casos')
plt.bar(x + width/2, monthly_data['deaths'], width, label='Muertes')

# Etiquetas y formato
plt.xlabel('Mes')
plt.ylabel('Cantidad')
plt.title('Casos y muertes por COVID-19 por mes')
plt.xticks(ticks=x[::3], labels=monthly_data['year_month'][::3], rotation=45)  # muestra cada 3er mes para claridad
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Creamos un nuevo DataFrame para almacenar las estadísticas COVID-19 - De barras apiladas - mes a mes juntando graficos
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Calculamos estadísticas descriptivas.
media = df_sp500['Cerrar/último'].mean()
mediana = df_sp500['Cerrar/último'].median()
moda = df_sp500['Cerrar/último'].mode()[0]

# Creamos la figura con dos subgraficos (uno para cada grafico)
fig, axes = plt.subplots(2, 1, figsize=(14, 14))

# Grafico 1: Precio de Cierre del S&P 500
axes[0].plot(df_sp500.index, df_sp500['Cerrar/último'], label='Precio de Cierre', color='b')
axes[0].axhline(media, color='r', linestyle='--', label=f'Media: {media:.2f}')
axes[0].axhline(mediana, color='g', linestyle='--', label=f'Mediana: {mediana:.2f}')
axes[0].axhline(moda, color='y', linestyle='--', label=f'Moda: {moda:.2f}')
axes[0].set_title('Precio de Cierre del S&P 500 con Media, Mediana y Moda')
axes[0].set_xlabel('Fecha')
axes[0].set_ylabel('Precio de Cierre')
axes[0].legend()
axes[0].grid(True)

# Cargar los datos de COVID-19
df_covid = pd.read_csv('covid19.csv')

# Asegurarse de que la columna 'date' esté en formato datetime
# Ensure the 'date' column is correctly accessed in df_covid
if 'date' in df_covid.columns:
	df_covid['date'] = pd.to_datetime(df_covid['date'])
else:
	raise KeyError("The 'date' column is missing in df_covid.")

# Crear columna de año-mes
df_covid['year_month'] = df_covid['date'].dt.to_period('M')

# Agrupar por año-mes y sumar los casos y muertes
monthly_data = df_covid.groupby('year_month')[['cases', 'deaths']].sum().reset_index()
monthly_data['year_month'] = monthly_data['year_month'].astype(str)

# Grafico 2: Casos y muertes por COVID-19 por mes
x = np.arange(len(monthly_data['year_month']))
width = 0.4

axes[1].bar(x - width/2, monthly_data['cases'], width, label='Casos')
axes[1].bar(x + width/2, monthly_data['deaths'], width, label='Muertes')
axes[1].set_xlabel('Mes')
axes[1].set_ylabel('Cantidad')
axes[1].set_title('Casos y muertes por COVID-19 por mes')
axes[1].set_xticks(x[::3])
axes[1].set_xticklabels(monthly_data['year_month'][::3], rotation=45)  # Muestra cada 3er mes
axes[1].legend()
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

# Ajustamos
plt.tight_layout()

# Mostramos el grafico combinado
plt.show()

# Imprimir las estadísticas
print(f"Media del precio de cierre: {media}")
print(f"Mediana del precio de cierre: {mediana}")
print(f"Moda del precio de cierre: {moda}")

```

A continuacion mostramos el grafico de los ultimos 10 años del valor de NETFLIX.Observamos que el valor durante el comienzo del año 2020, inicio de la pandemia,  no realizo una fuerte correccion, sino todo lo contrario.Ello es debido que durante la pantdemia la poblacion tuvo que estar confinada en casa y se realizaron millones de suscripciones a la plataforma.
A medida que las restricciones de movimiento eran menores y las vacunas estaban resolviendo en parte la pandemia las suscripciones fueron bajando en picado por el consiguioente descenso del valor en bolsa.


```python
# Ahora vamos a graficar el precio de cierre de Netflix
# Leemos el archivo CSV de Netflix
ruta_netflix_csv = 'netflixhistorico.csv'
data = pd.read_csv(ruta_netflix_csv, sep=',')

# Limpiamos la columna 'Cerrar/último' para que sea numerica eliminando el simbolo '$' y las comas.
data['Cerrar/último'] = data['Cerrar/último'].replace({'\\$': '', ',': ''}, regex=True).astype(float)

# Convertimos la columna 'Fecha' en formato datetime
data['Fecha'] = pd.to_datetime(data['Fecha'], format='%m/%d/%Y')

# Creamos el grafico de evolucion del precio de cierre
plt.figure(figsize=(10,6))
plt.plot(data['Fecha'], data['Cerrar/último'], label='Precio de Cierre', color='blue')

# Añadimos título y etiquetas
plt.title('Evolucion del Precio de Cierre de Netflix')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargamos el archivo CSV
df = pd.read_csv('zoom.csv')

# Convertir la columna 'Date' a formato datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Limpiamos los valores de la columna 'Close/Last' eliminando el signo '$' y convirtiendo a float para graficar
df['Close/Last'] = df['Close/Last'].replace({'\\$': '', ',': ''}, regex=True).astype(float)

# Crear el gráfico
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Close/Last'], label='Precio de Cierre', color='blue')

# Personalizar el gráfico
plt.title('Precio de Cierre de Zoom')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre en dólares')
plt.xticks(rotation=45)  # Rotamos las etiquetas de fecha para mejorar la visibilidad
plt.legend()

# Mostrar el gráfico
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Cargamos el archivo CSV
file_path = 'petroleo.csv'
data = pd.read_csv(file_path)

# Convertimos la columna 'Fecha' a tipo de dato datetime para facilitar el gráfico
data['Fecha'] = pd.to_datetime(data['Fecha'], format='%d.%m.%Y')

# Aseguramos de que la columna 'Último' sea numerica
data['Último'] = data['Último'].str.replace(',', '.').astype(float)

# Creamos el gráfico
plt.figure(figsize=(10, 6))
plt.plot(data['Fecha'], data['Último'], marker='*', linestyle='-', color='g')

# Etiquetas y título
plt.title('Evolucion del precio del petroleo')
plt.xlabel('Fecha')
plt.ylabel('Precio en dolares')
plt.xticks(rotation=45)
plt.grid(True)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
